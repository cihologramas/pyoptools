```python
## Load the new catalog on top of the old one

d = zmf2dict("THORLABS.ZMF")

import configparser

components = configparser.ConfigParser()
components.read("thorlabs_spherical.cmp")

for k in d:
    if k in ["ACA254-060-B"]:  # Lenses that seem to have errors in the library
        continue

    z = zmx2pyoptoolsSP(d, k)
    if z is not None:
        components[k] = z[k]

## save the new library
with open("thorlabs_spherical.cmp", "w") as configfile:
    components.write(configfile)
```

```python
# Import zmx files in the files subfolder files

components = configparser.ConfigParser()
components.read("thorlabs_spherical.cmp")
import os

file_list = []
for file in os.listdir("./files"):
    if file.endswith(".zmx"):
        file_list.append(file)

for file in file_list:
    with open(os.path.join("./files/", file), "rb") as f:
        fc = f.read()
        c = fc[2:].decode("latin").replace("\0", "")
    key = file.replace("-Zemax(ZMX).zmx", "")
    z = zmx2pyoptoolsSP({key: c}, key)
    if z is not None:
        components[key] = z[key]

with open("thorlabs_spherical.cmp", "w") as configfile:
    components.write(configfile)
```
