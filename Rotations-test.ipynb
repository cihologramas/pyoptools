# Testing rotations

```python
from pyoptools.all import *
```

```python
# Testing first rotation tuple

S=System()

R=[Ray(pos=(0,0,0),dir=(1,0,0),wavelength=.450), # Blue
   Ray(pos=(0,0,0),dir=(0,1,0),wavelength=.550), # Yellow
   Ray(pos=(0,0,0),dir=(0,0,1),wavelength=.650)] # Red



S.ray_add(R)
S.propagate()

Plot3D(S,center=(0,0,0),size=(4,3),scale=150,rot=[(0,0,pi/2)])
```

```python
# Testing second rotation tuple

S=System()

R=[Ray(pos=(0,0,0),dir=(1,0,0),wavelength=.450), # Blue
   Ray(pos=(0,0,0),dir=(0,1,0),wavelength=.550), # Yellow
   Ray(pos=(0,0,0),dir=(0,0,1),wavelength=.700)] # Red

S.ray_add(R)
S.propagate()

Plot3D(S,center=(0,0,0),size=(4,3),scale=150,rot=[(0,0,pi/2),(pi/2,0,0)])
```

```python
# Testing other rotation tuples

S=System()

R=[Ray(pos=(0,0,0),dir=(1,0,0),wavelength=.450), # Blue
   Ray(pos=(0,0,0),dir=(0,1,0),wavelength=.550), # Yellow
   Ray(pos=(0,0,0),dir=(0,0,1),wavelength=.700)] # Red

S.ray_add(R)
S.propagate()

Plot3D(S,center=(0,0,0),size=(4,3),scale=150,rot=[(0,0,pi/4),(pi/2,0,0),(0,0,pi/4),(0,0,0),(0,0,0)])

# Rotations just add in the original coordinate system
```
